## A. Importar librerías requeridas

En la siguiente casilla importaremos los módulos necesarios para crear las visualizaciones

```python
# Importando librerías
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

## B. Cargar el dataset
El dataset que cargaremos consiste en propinas dadas en un restaurant

```python
tips = sns.load_dataset('tips')
```

```python
# Observaciones en el dataset
len(tips)
```

```python
# Revisemos las 5 primeras filas
tips.head()
```

## C. Histograma

```python
# Veremos la distribución de la variable tip (propina)
# con un histograma

ax = sns.distplot(tips.tip)
```

Por defecto vemos una curva que aproxima una función de densidad de probabilidad. Si queremos omitirla usamos el atributo **kde = False**. También podemos manipular el **color** y el número de **bins**.

```python
sns.set()
sns.set_style('whitegrid')
ax = sns.distplot(tips.tip, kde= False, color='red', bins=25)
```

Otra opción interesante es poner varios histogramas en el mismo gráfico; por ejemplo para comparar la distribución de propinas en la cena versus el almuerzo. Nota: No debe parecer extraño el hecho que se use en el siguiente código tanto sns y plt (pyplot) para el mismo gráfico, ya que seaborn utiliza internamente matplotlib.

```python
sns.distplot( tips[tips.time == "Lunch"].tip, kde= False,
            color="red", label="Propinas en Almuerzo")
sns.distplot( tips[tips.time =="Dinner"].tip, kde= False,
            color="skyblue", label="Propinas en Cena")
plt.legend()
plt.show()
```

Finalmente, podríamos querer poner estos histogramas pero en gráficos uno al lado del otro.

```python
f, axes = plt.subplots(1,2, figsize=(7,4), sharex=True)
sns.distplot( tips[tips.time == "Lunch"].tip, kde= False,
            color="red", label="Propinas en Almuerzo", ax=axes[0])
sns.distplot( tips[tips.time =="Dinner"].tip, kde= False,
            color="skyblue", label="Propinas en Cena", ax=axes[1])
```

## D. Gráfico de caja (Boxplot)
Haremos un gráfico de caja con la información de propinas para comparar la información que nosm uestra respecto a un histograma. Recuerde que la línea dentro de la caja es la mediana.

```python
# Veremos la distribución de la variable tip (propina)
# con un gráfico de cajas

ax = sns.boxplot(tips.tip)
```

```python
# Para forzar que quede vertical y cambiarle el color:

ax = sns.boxplot(tips.tip, orient="v", color="red")
```

```python
# Ahora veamos si cambian según el día de la semana, y cambiemos 
# la plaeta de colores por defecto

sns.boxplot(x="day", y="tip", data=tips, palette="Set3")
```

## E. Gráfico de Violín

Recordemos que el gráfico de violín es similar al boxplot, pero además agrega una estimación de densidad de probabilidad para visualizar mejor la distribución de datos

```python
# Distribución de las cuentas en el restaurant con gráfico de violín

ax = sns.violinplot(x=tips['total_bill'])
```

```python
# Revisemos ahora cómo cambia por día de la semana
ax = sns.violinplot(x="day", y="total_bill", data=tips)
```

```python
# Podemos además separar estos gráficos en un segundo nivel,
# dependiendo de si lamesa es para fumadores o no

ax = sns.violinplot(x="day", y="total_bill", hue="smoker",
                   data=tips, palette="muted")
```
